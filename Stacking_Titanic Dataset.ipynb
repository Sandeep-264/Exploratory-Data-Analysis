```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from mlxtend.classifier import StackingClassifier
```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('https://www.openml.org/data/get_csv/16826755/phpMYEkMl')
```

```python
# Drop unnecessary columns
titanic_df.drop(['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest'], axis=1, inplace=True)
```

```python
# Convert categorical variables to numerical using one-hot encoding
titanic_df = pd.get_dummies(titanic_df)
```

```python
# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(titanic_df.drop('survived', axis=1),
                                                    titanic_df['survived'],
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
# Create the base models
models = [
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42)),
    ('ada', AdaBoostClassifier(n_estimators=100, random_state=42)),
    ('gbc', GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ('svc', SVC(kernel='linear', probability=True, random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42))
]
```

```python
# Create the meta-model
meta_model = LogisticRegression()
```

```python
# Create the stacking classifier
stacking = StackingClassifier(classifiers=[model[1] for model in models],
                              meta_classifier=meta_model)
```

```python
# Fit the stacking classifier on the randomized and distributed training datasets
stacking.fit(X_train, y_train)
```

```python
# Predict on the test dataset
y_pred = stacking.predict(X_test)
```

```python
# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

# BOOSTING

```python
import pandas as pd
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load the Titanic dataset
titanic_df = pd.read_csv('https://www.openml.org/data/get_csv/16826755/phpMYEkMl')

# Drop unnecessary columns
titanic_df.drop(['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest'], axis=1, inplace=True)

# Convert categorical variables to numerical using one-hot encoding
titanic_df = pd.get_dummies(titanic_df)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(titanic_df.drop('survived', axis=1),
                                                    titanic_df['survived'],
                                                    test_size=0.3,
                                                    random_state=42)

# Create the base model
base_model = DecisionTreeClassifier(max_depth=1, random_state=42)

# Create the ensemble model
ensemble_model = AdaBoostClassifier(base_model, n_estimators=100, random_state=42)

# Fit the ensemble model on the randomized and distributed training datasets
ensemble_model.fit(X_train, y_train)
```

```python
# Predict on the test dataset
y_pred = ensemble_model.predict(X_test)
```

```python
# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
